```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import LocalOutlierFactor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.cluster import KMeans
from sklearn.ensemble import GradientBoostingClassifier
```

```python
df = pd.read_csv('/kaggle/input/sml-project/train.csv')
df
```

```python
data = df.to_numpy()
# X->data and Y->target values of our data
X = data[:, 1:-1]
Y = data[:,-1]

X.shape
```

```python
lof = LocalOutlierFactor(n_neighbors=27)
lof.fit(X)

lof_scores = lof.negative_outlier_factor_

outliers = np.where(lof_scores < -1.32)[0]

X = np.delete(X, outliers, axis=0)
Y = np.delete(Y,outliers,axis=0)
print("Cleaned data shape:", X.shape)
print("Cleaned target shape:", Y.shape)
```

```python

# Standardize the data
scaler = StandardScaler()
# X = scaler.fit_transform(X)

# # Applying PCA
pca = PCA(n_components=450,random_state=7)
X = pca.fit_transform(X)

# Applying LDA
lda = LinearDiscriminantAnalysis(n_components=19)

# # # # Fit the LDA model to the data
X = lda.fit_transform(X, Y)
X

```

```python
print("Reduced data shape:", X.shape)
print("Reduces target shape:", Y.shape)

```

```python
# Applying clustering Algorithm

kmeans = KMeans(n_clusters=20)
kmeans.fit(X)
cluster_assignments = kmeans.labels_

print("Unique clusters - > ",np.unique(cluster_assignments))
cluster_assignments =scaler.fit_transform(cluster_assignments.reshape(-1,1)) 
X = np.concatenate((X, cluster_assignments), axis=1)

X
```

```python
X.shape
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2,random_state=52)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

k = 3

# Create a k-fold cross-validation object
kf = KFold(n_splits=k, shuffle=True, random_state=42)
```

```python
clf = LogisticRegression(multi_class='ovr', solver='lbfgs',max_iter=1000)
scores = cross_val_score(clf, X_train, y_train, cv=kf)

# Print the cross-validation scores
print("Cross-validation scores:", scores)
print("Mean score:", scores.mean())
```

```python


clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
```
